```python
from Bio.PDB import PDBParser, PDBIO

# Parse the PDB file
parser = PDBParser(QUIET=True)
structure = parser.get_structure("7o7k", "7o7k.pdb")

# Function to check if a residue is a ligand
def is_ligand(residue):
    # Standard residues have a blank insertion code (" ") in their ID
    # Non-standard residues are ligands
    return residue.id[0] != " "

# Remove ligands from the structure
for model in structure:
    for chain in list(model):  # Use list() to allow modification during iteration
        for residue in list(chain):
            if is_ligand(residue):
                print(f"Removing ligand: {residue.resname}, Residue ID: {residue.id}")
                chain.detach_child(residue.id)

# Save the modified structure
io = PDBIO()
io.set_structure(structure)
io.save("7o7k_no_ligands.pdb")
```
